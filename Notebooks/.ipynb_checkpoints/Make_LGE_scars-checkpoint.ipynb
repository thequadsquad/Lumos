```python
def add_fwhm_scar(img, anno):
    if not anno.has_contour('lge_ref') or not anno.has_contour('lv_myo'): return anno
    lge_ref_pixels = anno.get_pixel_values('lge_ref', img)
    myo_pixels     = anno.get_pixel_values('lv_myo',  img)
    enhance_mean = np.nanmean(lge_ref_pixels)
    myo_mean     = np.nanmean(myo_pixels)
    fwhm_val     = myo_mean + (enhance_mean - myo_mean)/2.0
    # make scar mask
    scar_mask = img >= fwhm_val
    myo_mask  = anno.get_cont_as_mask('lv_myo')
    exc_mask  = anno.get_cont_as_mask('lge_ex')
    scar_mask = scar_mask * myo_mask * (1-exc_mask)
    poly = utils.to_polygon(scar_mask)
    anno.anno['fwhm_scar'] = dict()
    anno.anno['fwhm_scar']['subpixelResolution'] = 1
    anno.anno['fwhm_scar']['imageSize'] = anno.anno['lv_myo']['imageSize']
    anno.anno['fwhm_scar']['pixelSize'] = anno.anno['lv_myo']['pixelSize']
    anno.anno['fwhm_scar']['cont'] = poly
    anno.anno['fwhm_scar']['contType'] = 'FREE'
    return anno

def add_fwhm_scar_from_other_slice(img, anno, other_img, other_anno):
    # get threshold from others, use it for img and anno
    lge_ref_pixels = other_anno.get_pixel_values('lge_ref', img)
    myo_pixels     = other_anno.get_pixel_values('lv_myo',  img)
    enhance_mean = np.nanmean(lge_ref_pixels)
    myo_mean     = np.nanmean(myo_pixels)
    fwhm_val     = myo_mean + (enhance_mean - myo_mean)/2.0
    # make scar mask
    scar_mask = img >= fwhm_val
    myo_mask  = anno.get_cont_as_mask('lv_myo')
    exc_mask  = anno.get_cont_as_mask('lge_ex')
    scar_mask = scar_mask * myo_mask * (1-exc_mask)
    poly = utils.to_polygon(scar_mask)
    anno.anno['fwhm_scar'] = dict()
    anno.anno['fwhm_scar']['subpixelResolution'] = 1
    anno.anno['fwhm_scar']['imageSize'] = anno.anno['lv_myo']['imageSize']
    anno.anno['fwhm_scar']['pixelSize'] = anno.anno['lv_myo']['pixelSize']
    anno.anno['fwhm_scar']['cont'] = poly
    anno.anno['fwhm_scar']['contType'] = 'FREE'
    return anno

def get_anno_with_scar_fwhm(imgo, annos, slice_nr):
    img  = imgo.get_img(slice_nr, 0)
    try: anno = annos[imgo.depthandtime2sop[(slice_nr,0)]]
    except: return
    if anno.has_contour('lge_ref'): 
        print('1')
        return add_fwhm_scar(img, anno)
    print('2')
    for i in range(imgo.nr_slices):
        low_d  = max(0, slice_nr - i)
        high_d = min(slice_nr + i, self.nr_slices-1)
        for d in [low_d, high_d]:
            other_anno = self.get_anno(d, 0)
            if other_anno.has_contour('lge_ref'):
                other_img = self.get_img(d, 0)
                return anno.add_fwhm_scar_from_other_slice(img, anno, other_img, other_anno)


"""
def get_anno_with_scar_fwhm(self, slice_nr, exclude=True):
    img  = self.get_img (slice_nr, 0)
    anno = self.get_anno(slice_nr, 0)
    if anno.has_contour('saEnhancementReferenceMyoContour'):
        anno.add_fwhm_scar(img, exclude=exclude)
        return anno
    for i in range(self.nr_slices):
        low_d  = max(0, slice_nr - i)
        high_d = min(slice_nr + i, self.nr_slices-1)
        for d in [low_d, high_d]:
            other_anno = self.get_anno(d, 0)
            if other_anno.has_contour('saEnhancementReferenceMyoContour'):
                other_img = self.get_img(d, 0)
                other_anno.add_fwhm_scar(other_img, exclude=exclude)
                anno.add_fwhm_scar_other_slice(img, other_anno, exclude=exclude)
                return anno

def preprocess_scars(self):
    for d in range(self.nr_slices):
        anno = self.get_anno(d,0)
        if anno.sop not in self.case.annos_sop2filepath: continue
        new_anno = anno.anno
        for exclude in [False, True]:
            anno = self.get_anno_with_scar_fwhm(d, exclude)
            new_anno = {**new_anno, **anno.anno}
"""
```
