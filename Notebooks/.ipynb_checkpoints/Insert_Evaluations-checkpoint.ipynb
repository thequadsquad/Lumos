```python
from time import time
from pandas import DataFrame
import traceback

from bson.objectid import ObjectId
from RoomOfRequirement.Quad import QUAD_Manager
from RoomOfRequirement.Evaluation import *
```

```python
quad = QUAD_Manager()
```

```python
# Cohorts
cohorts = list(quad.coho_coll.find())
for c in cohorts: print(c['name']+' '*(15-len(c['name'])), c['ownername'], len(c['studyuids']), c['_id'])
print()

# Tasks
tasks = list(quad.task_coll.find())
for t in tasks: print(t['displayname'], ',\t Nr Studyuids: ', len(t['studyuids']), ',\tID: ', t['_id'])
```

```python
# Select cohort and tasks
cohort = quad.coho_coll.find_one({'_id': ObjectId('661f9e3ffa2b5fc9df8d02ae')}) # Debugging Cases
task1  = quad.task_coll.find_one({'_id': ObjectId('6644b38bebda13c0a882a9fe')}) # NI
task2  = quad.task_coll.find_one({'_id': ObjectId('6644b078ebda13c0a882a9fd')}) # AI
```

```python
imgtypes  = list(quad.imgo_coll.find({'studyuid' : {'$in': cohort['studyuids']}}).distinct('imagetype'))
if 'LAX CINE 3CV' in imgtypes: imgtypes.remove('LAX CINE 3CV')
if 'LAX 3CV' in imgtypes: imgtypes.remove('LAX 3CV')
print(imgtypes)
print(len(cohort['studyuids']))
```

```python
suids = cohort['studyuids']
for i, suid in enumerate(suids):
    case = quad.case_coll.find_one({'studyuid': suid})
    print('\n', i+1, ' of ', len(suids), '\t', case['name'], '\t', suid, '\n')
    for t in [task1, task2]: 
        print('\n\tTask: ', t['displayname'], t['_id'])
        for imgtype in imgtypes:
            print('\tImagetype: ', imgtype)
            imgo = quad.imgo_coll.find_one({'studyuid': suid, 'imagetype': imgtype})
            print('\t\tImage organizer exists? ', imgo is not None)
            if imgo is not None:
                sops = list(imgo['sop2depthandtime'].keys())
                anno = quad.anno_coll.find_one({'studyuid': suid, 'task_id': t['_id'], 'sop': {'$in': sops}})
                print('\t\tTask has at least one anno for image organizer? ', anno is not None)
            if imgo is None or anno is None: continue
            try: eva = Evaluation(quad, task_id=t['_id'], imagetype=imgtype, studyuid=suid); print('\t\tInit Succeeded')
            except: print('\t\tInitialization Failed: \n', traceback.format_exc()); continue
            try: eva.evaluate(); print('\t\tCalculation Succeeded')
            except: print('\t\tCalculation Failed: \n', traceback.format_exc()); continue
            try:
                eva_dict = eva.__dict__
                eva_dict.pop('db'); eva_dict.pop('imgo'); eva_dict.pop('depthandtime2sop')
                quad.eval_coll.insert_one(eva_dict)
                print('\t\tInsertion Succeeded')
            except: print('\t\tInsertion Failed: \n', traceback.format_exc()); continue
```
