```python
from time import time
import os
from pandas import DataFrame

from bson.objectid import ObjectId
from RoomOfRequirement.Quad import QUAD_Manager
```

```python
quad = QUAD_Manager()
```

```python
# Cohorts
cohorts = list(quad.coho_coll.find())
for c in cohorts: print(c['name']+' '*(15-len(c['name'])), c['ownername'], len(c['studyuids']), c['_id'])
print()

# Tasks
tasks = list(quad.task_coll.find())
for t in tasks: print(t['displayname'], ',\t Nr Studyuids: ', len(t['studyuids']), ',\tID: ', t['_id'])
```

```python
task1  = quad.task_coll.find_one({'_id': ObjectId('6644b38bebda13c0a882a9fe')})
task2  = quad.task_coll.find_one({'_id': ObjectId('6644b078ebda13c0a882a9fd')})

for cohort in list(quad.coho_coll.find()):
    imgtypes  = list(quad.imgo_coll.find({'studyuid' : {'$in': cohort['studyuids']}}).distinct('imagetype'))
    imgos     = list(quad.imgo_coll.find({'studyuid' : {'$in': cohort['studyuids']}}))
    suid2imgo = dict()
    for imgo in imgos:
        imgtype, suid, sops = imgo['imagetype'], imgo['studyuid'], list(imgo['sop2depthandtime'].keys())
        if suid not in suid2imgo.keys(): suid2imgo[suid] = dict(); suid2imgo[suid]['name']=imgo['name']
        suid2imgo[suid][imgtype] = [suid2imgo[suid]['name'], True, len(sops)]
        # task 1
        nr_annos = quad.anno_coll.count_documents({'studyuid': suid, 'sop': {'$in': sops}, 'task_id': task1['_id']})
        suid2imgo[suid][imgtype].append(nr_annos)
        evaluation = quad.eval_coll.find_one({'studyuid': suid, 'imagetype': imgtype, 'task_id': task1['_id']})
        suid2imgo[suid][imgtype].append(bool(evaluation))
        # task 2
        nr_annos = quad.anno_coll.count_documents({'studyuid': suid, 'sop': {'$in': sops}, 'task_id': task2['_id']})
        suid2imgo[suid][imgtype].append(nr_annos)
        evaluation = quad.eval_coll.find_one({'studyuid': suid, 'imagetype': imgtype, 'task_id': task2['_id']})
        suid2imgo[suid][imgtype].append(bool(evaluation))

    cols = [] 
    for imgtype in imgtypes: cols += ['Studyname', imgtype, 'Nr Images', 'Nr Annos - '+task1['displayname'], 'Has Evaluation - '+task1['displayname'], 'Nr Annos - '+task2['displayname'], 'Has Evaluation - '+task2['displayname']]
    cols += ['Studyuid', 'TaskID - '+task1['displayname'], 'TaskID - '+task2['displayname']]
    
    rows = []
    for suid in suid2imgo.keys():
        row = []
        for i, imgtype in enumerate(imgtypes):
            if imgtype not in suid2imgo[suid].keys(): row.extend([suid2imgo[suid]['name'], '--', '--', '--', '--', '--', '--'])
            else:                                     row.extend(suid2imgo[suid][imgtype])
        row.extend([suid, task1['_id'], task2['_id']])
        rows.append(row)

    df = DataFrame(rows, columns=cols)
    #df.to_csv('/Users/thomas/Desktop/out_example.csv')
    df.to_excel(os.path.join('C:/Users/AG-CMR/Desktop/NickPapers_Quad/Debugging_Support/Has_Images_Has_Annotations_21052024', cohort['name']+'_'+task1['displayname']+'_'+task2['displayname']+'.xlsx'))
```
