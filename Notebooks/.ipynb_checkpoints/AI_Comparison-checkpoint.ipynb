```python
tasks = list(quad.task_coll.find())
for t in tasks:
    print(t['displayname'])
evals_gold = list(quad.case_coll.find())
```

```python
evals_gold = quad.eval_coll.find({'task_id': tasks[0]['_id'], 'imagetype': 'SAX CINE'})
evals_gold = [Evaluation(quad, task_id=e['task_id'], studyuid=e['studyuid'], imagetype='SAX CINE') for e in evals_gold]
evals_gold = {e.studyuid: e for e in evals_gold}

evals_AI_1 = quad.eval_coll.find({'task_id': tasks[1]['_id'], 'imagetype': 'SAX CINE'})
evals_AI_1 = [Evaluation(quad, task_id=e['task_id'], studyuid=e['studyuid'], imagetype='SAX CINE') for e in evals_AI_1]
evals_AI_1 = {e.studyuid: e for e in evals_AI_1}

evals_AI_2 = quad.eval_coll.find({'task_id': tasks[2]['_id'], 'imagetype': 'SAX CINE'})
evals_AI_2 = [Evaluation(quad, task_id=e['task_id'], studyuid=e['studyuid'], imagetype='SAX CINE') for e in evals_AI_2]
evals_AI_2 = {e.studyuid: e for e in evals_AI_2}

evals_AI_3 = quad.eval_coll.find({'task_id': tasks[3]['_id'], 'imagetype': 'SAX CINE'})
evals_AI_3 = [Evaluation(quad, task_id=e['task_id'], studyuid=e['studyuid'], imagetype='SAX CINE') for e in evals_AI_3]
evals_AI_3 = {e.studyuid: e for e in evals_AI_3}
```

```python
print(len(evals_gold))
print(len(evals_AI_1))
print(len(evals_AI_2))
print(len(evals_AI_3))
```

```python
suids1 = set(evals_gold.keys())
suids2 = set(evals_AI_1.keys())
suids3 = set(evals_AI_2.keys())
suids4 = set(evals_AI_3.keys())

common_suids  = sorted(list(suids1.intersection(suids2).intersection(suids3).intersection(suids4)))
ill_suids     = []
healthy_suids = [] 
cases         = list(quad.case_coll.find({'studyuid': {'$in': common_suids}}))
sgs_suids     = quad.coho_coll.find_one({'name': 'SGS'})['studyuids']
for c in cases:
    if   'primary_disease' not in c.keys(): healthy_suids.append(c['studyuid'])
    elif 'Hypertrophic Cardiomyopathy' in c['primary_disease']: ill_suids.append(c['studyuid'])
    elif 'Dilatative Cardiomyopathy' in c['primary_disease']:   ill_suids.append(c['studyuid'])
    elif 'Systemic Disease' in c['primary_disease']:            ill_suids.append(c['studyuid'])
    elif c['studyuid'] in sgs_suids:                            ill_suids.append(c['studyuid'])
    else: healthy_suids.append(c['studyuid'])

print('All: ', len(common_suids))
print('Changed: ', len(ill_suids))
print('Unchanged: ', len(common_suids) - len(ill_suids))
```

```python
def get_df(df, ai_nr, health_st=None):
    if health_st is None: return np.asarray(df[df['AI']==ai_nr]['Parameter Difference'])
    return np.asarray(df[(df['AI']==ai_nr) & (df['Type']==health_st)]['Parameter Difference'])

cpnames = ['LVESV', 'LVEDV', 'LVSV', 'LVEF', 'LVM', 'RVESV', 'RVEDV', 'RVSV', 'RVEF']
for cp_i, cpn in enumerate(cpnames):
    df = DataFrame(rows_by_parameters[cpn], columns=['Type', 'AI', 'Parameter Difference', 'Parameter Average'])
    print(cpn, end=', ')
    print("%.1f" % np.nanmean(get_df(df, 'AI 1', 'Unchanged'))+'  ' + 
          "%.1f" % np.nanstd (get_df(df, 'AI 1', 'Unchanged')), 
          "%.1f" % np.nanmean(get_df(df, 'AI 1', 'Changed'))+'  ' + 
          "%.1f" % np.nanstd (get_df(df, 'AI 1', 'Changed')), 
          "%.1f" % np.nanmean(get_df(df, 'AI 2', 'Unchanged'))+'  ' + 
          "%.1f" % np.nanstd (get_df(df, 'AI 2', 'Unchanged')),
          "%.1f" % np.nanmean(get_df(df, 'AI 2', 'Changed'))+'  ' + 
          "%.1f" % np.nanstd (get_df(df, 'AI 2', 'Changed')),
          "%.1f" % np.nanmean(get_df(df, 'AI 3', 'Unchanged'))+'  ' + 
          "%.1f" % np.nanstd (get_df(df, 'AI 3', 'Unchanged')),
          "%.1f" % np.nanmean(get_df(df, 'AI 3', 'Changed'))+'  ' + 
          "%.1f" % np.nanstd (get_df(df, 'AI 3', 'Changed')))

print()
for cp_i, cpn in enumerate(cpnames):
    df = DataFrame(rows_by_parameters[cpn], columns=['Type', 'AI', 'Parameter Difference', 'Parameter Average'])
    print(cpn, end=', ')
    print("%.1f" % np.nanmean(get_df(df, 'AI 1', None)), 
          "%.1f" % np.nanmean(get_df(df, 'AI 2', None)),
          "%.1f" % np.nanmean(get_df(df, 'AI 3', None)))
```

```python
def calculate_segmentation_metrics(suids, evas, ai_evas):
    dice = DiceMetric()
    hd   = HausdorffMetric()
    
    lv_endo_dices_both, lv_myo_dices_both, rv_endo_dices_both = [], [], []
    lv_endo_dices, lv_myo_dices, rv_endo_dices                = [], [], []
    lv_endo_hds,   lv_myo_hds,   rv_endo_hds                  = [], [], []
    
    for ii, suid in enumerate(suids):
        if ii%10==0: print(ii, ' of ', len(suids))
        e1, e2 = evas[suid], ai_evas[suid]
        dcm = e1.get_dcm(0,0)
        for d in range(e1.nr_slices):
            lvesp = cps['LVESP'].get_val(e1)
            anno1, anno2 = e1.get_anno(d,lvesp), e2.get_anno(d,lvesp)
            lv_endo_dices.append(dice.get_val(anno1.get_contour('lv_endo'), anno2.get_contour('lv_endo')))
            if anno1.has_contour('lv_endo') and anno2.has_contour('lv_endo'):
                lv_endo_dices_both.append(dice.get_val(anno1.get_contour('lv_endo'), anno2.get_contour('lv_endo')))
                lv_endo_hds.append(hd.get_val(anno1.get_contour('lv_endo'), anno2.get_contour('lv_endo'), dcm))
            
            lvedp = cps['LVEDP'].get_val(e1)
            anno1, anno2 = e1.get_anno(d,lvedp), e2.get_anno(d,lvedp)
            lv_endo_dices.append(dice.get_val(anno1.get_contour('lv_endo'), anno2.get_contour('lv_endo')))
            lv_myo_dices .append(dice.get_val(anno1.get_contour('lv_myo'), anno2.get_contour('lv_myo')))
            if anno1.has_contour('lv_endo') and anno2.has_contour('lv_endo'):
                lv_endo_dices_both.append(dice.get_val(anno1.get_contour('lv_endo'), anno2.get_contour('lv_endo')))
                lv_endo_hds       .append(hd  .get_val(anno1.get_contour('lv_endo'), anno2.get_contour('lv_endo'), dcm))
            if anno1.has_contour('lv_myo') and anno2.has_contour('lv_myo'):
                lv_myo_dices_both.append(dice.get_val(anno1.get_contour('lv_myo'), anno2.get_contour('lv_myo')))
                lv_myo_hds       .append(hd  .get_val(anno1.get_contour('lv_myo'), anno2.get_contour('lv_myo'), dcm))

            rvesp = cps['RVESP'].get_val(e1)
            anno1, anno2 = e1.get_anno(d,rvesp), e2.get_anno(d,rvesp)
            rv_endo_dices.append(dice.get_val(anno1.get_contour('rv_endo'), anno2.get_contour('rv_endo')))
            if anno1.has_contour('rv_endo') and anno2.has_contour('rv_endo'):
                rv_endo_dices_both.append(dice.get_val(anno1.get_contour('rv_endo'), anno2.get_contour('rv_endo')))
                rv_endo_hds.append(hd.get_val(anno1.get_contour('rv_endo'), anno2.get_contour('rv_endo'), dcm))

            rvedp = cps['RVEDP'].get_val(e1)
            anno1, anno2 = e1.get_anno(d,rvedp), e2.get_anno(d,rvedp)
            rv_endo_dices.append(dice.get_val(anno1.get_contour('rv_endo'), anno2.get_contour('rv_endo')))
            if anno1.has_contour('rv_endo') and anno2.has_contour('rv_endo'):
                rv_endo_dices_both.append(dice.get_val(anno1.get_contour('rv_endo'), anno2.get_contour('rv_endo')))
                rv_endo_hds       .append(hd  .get_val(anno1.get_contour('rv_endo'), anno2.get_contour('rv_endo'), dcm))
        #if ii==10: break
    return lv_endo_dices_both, lv_myo_dices_both, rv_endo_dices_both, lv_endo_dices, lv_myo_dices, rv_endo_dices, lv_endo_hds, lv_myo_hds, rv_endo_hds
```

```python
cpnames = ['LVESV', 'LVEDV', 'LVSV', 'LVEF', 'LVM', 'RVESV', 'RVEDV', 'RVSV', 'RVEF']
cpname2values = {cpn:[] for cpn in cpnames}
for suid in evals_gold.keys():
    try: eva1, eva2 = evals_gold[suid], evals_AI_1[suid]
    except: continue
    for cpname in cpnames:
        cp = v.clinical_parameters[cpname]
        cpname2values[cpname].append(cp.get_val_diff(eva1, eva2))
#print(cpname2values)
#self.set_size_inches(w=15, h=15)
#axes = self.subplots(3,3)
fig, axes = plt.subplots(3,3, figsize=(15,15))
for ax_i in range(9):
    i, j = ax_i//3, ax_i%3
    ax = axes[i][j]
    #try:
    name = cpnames[ax_i]
    cp = v.clinical_parameters[name]
    print(cp)
    ax.set_title(name, fontsize=16)
    if np.isnan(np.nanmean(cpname2values[name])): ax.set_title('No '+name+' Values Available')
    alpha = 0.5# if with_swarmplot else 0.0
    ci = 1.96 * np.nanstd(cpname2values[name]) / np.sqrt(len(cpname2values[name]))
    ax.axhspan(-cp.tol_range, cp.tol_range, facecolor='0.6', alpha=0.5)
    #ax.errorbar([name], [np.nanmean(cpname2values[name])], yerr=ci, fmt ='o', c='r')
    df = DataFrame(cpname2values[name], columns=['Values'])
    #sns.pointplot(data=df, y=cpname2values[name], errorbar=("ci", 100), join=False, color='r')
    sns.swarmplot(ax=ax, y=cpname2values[name], palette=sns.color_palette("Blues")[4:], 
                  dodge=True, size=5, alpha=alpha)
    sns.pointplot(ax=ax, data=df, y=cpname2values[name], errorbar=("ci", 95), palette=sns.color_palette("Blues")[4:])
    maxx = np.max([np.abs(np.nanmin(cpname2values[name])), np.abs(np.nanmax(cpname2values[name])),
                   np.abs(np.nanmean(cpname2values[name])-ci), np.abs(np.nanmean(cpname2values[name])+ci), 
                   cp.tol_range])
    ax.set_ylim(ymin=-maxx-2, ymax=maxx+2)
    ax.set_ylabel(name + ' ' + cp.unit)
    ax.set_xlabel('')
    #ax.set_xticklabels(['Conf. Interval', 'Scatterplot'])
    ax.tick_params(axis='both', which='major', labelsize=14)
    #except Exception as e: print(e); continue
    break
fig.tight_layout()
```
